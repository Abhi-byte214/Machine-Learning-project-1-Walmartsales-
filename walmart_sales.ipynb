```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Walmart_Sales.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)
```

```python
print(df.head())
```

```python
X = df[['Temperature']]       
y = df['Weekly_Sales'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
m = model.coef_[0]
c = model.intercept_

print(f"Linear Regression Equation: Weekly_Sales = {m:.2f} × Temperature + {c:.2f}")
```

```python
y_pred = model.predict(X_test)
```

```python
sns.scatterplot(x=X_test['Temperature'], y=y_test, label='Actual Data')
```

```python
sns.lineplot(x=X_test['Temperature'], y=y_pred, color='red', label='Regression Line')

plt.xlabel("Temperature")
plt.ylabel("Weekly Sales")
plt.title("Temperature vs Weekly Sales (Linear Regression)")
plt.legend()

plt.show()
```

```python
user_temp = float(input("Enter Temperature value: "))
user_temp_df = pd.DataFrame({'Temperature': [user_temp]})

predicted_sales = model.predict(user_temp_df)
print(f"Predicted Weekly Sales: {predicted_sales[0]:.2f}")
```
